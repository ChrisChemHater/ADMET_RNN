```python
model_cv = train.ModelCV(PlusSolNet, folds=5, inner_hidden_size=150, feature_size=50, hidden_size=100)
model_cv.load_model('models/PlusSolNet/plus_solnet_model')
data_loader = PlusSolData('datasets/plus_esol_train.txt', batch_size=64)
Optimizer = torch.optim.Adam
optim_dict = {'lr': 1e-3}
loss = nn.MSELoss()
model_cv.train(data_loader, Optimizer, optim_dict, loss, early_stopping=True, tol=0.5)
model_cv.save_model('models/PlusSolNet/plus_solnet_model')
model_cv.predict(["CCCO", "c1ccc(N)cc1C=O"])
```
